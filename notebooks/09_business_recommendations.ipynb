**Author**: Nisarga N

**Dataset**: UCI Online Retail II  
**Objective**: Translate quantitative modeling results into strategic CRM guidelines.


## 1. Strategic Marketing Recommendations
We can leverage customer predicted value to adjust bidding rates and marketing allocation:

### 1.1 Customer Acquisition Cost (CAC) Caps
We should limit acquisition spend for new cohorts based on their predicted LTV:
- **High-Value Target Profile (Top 10% LTV)**: Allow up to **£150** CAC. Target channels: Google Search/Shopping for specific high-value categories, high-intent audiences.
- **Medium-Value Target Profile (Mid 40% LTV)**: Limit to **£40** CAC. Target channels: Instagram/Facebook lookalike audiences.
- **Low-Value Target Profile (Bottom 50% LTV)**: Limit to **£10** CAC. Target channels: Low-cost email/organic loops.

### 1.2 CRM & Loyalty Activation Campaigns
- **Champions Segment (VIP)**: Offer personal account managers, early product launches, and thank-you cards. Target ROI is high.
- **At-Risk Segment (Reactivation)**: Dispatch time-sensitive re-engagement discounts of 15% on categories they previously browsed.
- **Need Attention Segment (Nudge)**: Incentivize multi-buy offers to accelerate purchase frequency.
